```python
import csv
import pandas as pd
import numpy as np

import time
```

```python
def getCSVFileAsDataFrame(fileName):
    csvData = pd.read_csv(fileName)
    return pd.DataFrame(csvData)


def getCSVFileAsList(fileName):
    with open(fileName, 'r') as f:
        reader = csv.reader(f)
        return list(reader)


# def getCSVFileAsNpArray(fileName):
#     return np.genfromtxt(fileName, delimiter=',', skip_header=True)


# def selectRows(ls, index)
```

```python
df = getCSVFileAsDataFrame('NATMON_DATA_NATIONAL.csv')
ls = getCSVFileAsList('NATMON_DATA_NATIONAL.csv')
#arr = getCSVFileAsNpArray('NATMON_DATA_NATIONAL.csv')
```

```python
yearLs=[]
for row in ls:
    if row[2]=='2018':
        yearLs.append(row)
```

```python
labels = df['INDICATOR_ID'].drop_duplicates().values.tolist()
countrys = df['COUNTRY_ID'].drop_duplicates().values.tolist()
df = df[df['YEAR'] == 2018]
#data
```

```python
labelss = list(map(str, labels))
labelss=sorted(labelss)
countrys = sorted(countrys)
```

```python
# best performence
t1=time.time()
fullData=[]
for country in countrys:
    tmpData = []
    for row in yearLs:
        if row[1] == country:
            tmpData.append(row)
    countryData = [country]

    for label in labelss:
        flag = False
        for tmpRow in tmpData:
            # print(tmpRow[0], label)
            if tmpRow[0] == label:
                countryData.append(tmpRow[3])
                flag = True
                break
        if flag == False:
            countryData.append('NaN')
    fullData.append(countryData)
    print(country, len(countryData))
    # break
# print(len(fullData[0]), len(fullData[1]))
print((time.time()-t1))
```
